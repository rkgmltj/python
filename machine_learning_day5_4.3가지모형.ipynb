```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs
```

```python
x, y = make_blobs() # 실습용 데이터셋
x.shape, y.shape
```

```python
model = AgglomerativeClustering(n_clusters=3).fit(x)  # x를 3개의 그룹으로 모아
```

```python
p = model.fit_predict(x) # predict
```

```python
mglearn.discrete_scatter(x[:,0],x[:,1],p)
```

```python
n=7
model = AgglomerativeClustering(n_clusters=n).fit(x)  # x를 n개의 그룹으로 모아
p = model.fit_predict(x) # predict
mglearn.discrete_scatter(x[:,0],x[:,1],p)
```

```python
from sklearn.cluster import DBSCAN
```

```python
p = DBSCAN().fit_predict(x)  # k 개를 정해주지 않습니다.
```

```python
l = mglearn.discrete_scatter(x[:,0],x[:,1],p)
```

```python
# min_samples : 클러스터가 되기 위한 최소 샘플 수
# eps : 클러스터의 샘플들간의 거리
p = DBSCAN(eps=1.0).fit_predict(x)  # eps(샘플들간의 거리) 조정
l = mglearn.discrete_scatter(x[:,0],x[:,1],p)
```

```python
p = DBSCAN(eps=2.0).fit_predict(x)  # eps 조정
l = mglearn.discrete_scatter(x[:,0],x[:,1],p)
```

```python
p = DBSCAN(eps=1.0, min_samples=10).fit_predict(x)  # min_samples 조정
l = mglearn.discrete_scatter(x[:,0],x[:,1],p)
```
